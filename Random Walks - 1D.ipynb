```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
#Writing the random walk function

def RandomWalk(n):
    x = 0
    moves = [x]
    for i in range(n):
        dx = random.choice([-1, 1])
        x += dx
        moves.append(x)
    return x, moves

final_coord, move = RandomWalk(100)
print(final_coord, move)
```

```python
num_moves = range(len(move))
print(num_moves)
plt.plot(num_moves, move)
plt.axhline(np.mean(move), color='red')
plt.show()
```
