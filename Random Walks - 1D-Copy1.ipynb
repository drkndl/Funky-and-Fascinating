```python
import numpy as np
import matplotlib.pyplot as plt
import random
from IPython import display
```

```python
#Writing the random walk function

def RandomWalk(n, t):
    x = 0
    moves = [x]
    all_moves = []
    for j in range(t):
        moves = [0]
        x = 0
        for i in range(n):
            dx = random.choice([-1, 1])
            x += dx
            moves.append(x)
        all_moves.append(moves)
    return all_moves
```

```python
#Plotting the random walk over all iterations

def PlotWalk(n, t):
    arr = RandomWalk(n,t)
    for i in range(t):
        num_moves = range(len(arr[i]))
        plt.plot(num_moves, arr[i])
        plt.axhline(np.mean(arr[i]), color='red')
        display.clear_output(wait=True)
        display.display(plt.gcf())
        plt.pause(0.001)

PlotWalk(25, 10)
```
